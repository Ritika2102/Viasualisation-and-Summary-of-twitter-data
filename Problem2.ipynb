We set the variable search equal to the term we want to search and then use the method define by the twitteR package

```r
search <-"#Grammy"
searchResults <- searchTwitter(search , n = 2500)
```

We get the location of the user from its screenname and using LookupUsers method

```r
dataFrame <- twListToDF(searchResults)
userInfo <- lookupUsers(dataFrame$screenName)
userFrame <- twListToDF(userInfo)  # Convert then again to a data frame
write.csv(userFrame,file="DIC_Lab1.csv") #write it into a csv file
tweetFrame <- read.csv(file="DIC_Lab1.csv",header=TRUE,sep=",") #read from the csv file
loc1 <- na.omit(userFrame$location) # Then we take the location of the user from above data frame and put in a table
loc <- data.frame(loc1)
```

Convert the above data frame loc to a table so that we can get the frequency

```r
tab<-table(loc)
#tab
tweet_unique<-data.frame(location=names(tab),frequency=c(tab)) # we use the null function so we can clean the data
rownames(tweet_unique)<-NULL
tweet_freq<-tweet_unique  # we bind the tables so we can get address,lat,long and frequency all together in one table
tweet_freq <-cbind(tweet_freq, geocode(as.character(tweet_freq$location)))
tweet_freq <- na.omit(tweet_freq) # to omit the NA results
#tweet_freq 

```

Then we finally plot using the freq dataframe and size of the plot moves according to the frequency

```r
mapToPlot = get_map(zoom=4, scale=2, source = "google")
map.plot = ggmap(mapToPlot)
map.plot = map.plot + geom_point(data = tweet_freq, aes(x = lon, y = lat,size=frequency))
map.plot
```
